```python
# 引入 sentence-transformers 库中的 SentenceTransformer 类，用于加载预训练模型并生成句子嵌入向量。    
from sentence_transformers import SentenceTransformer

# 1. 加载一个预训练的句子嵌入模型，这里使用的是 'all-MiniLM-L6-v2'
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. 准备输入的句子，这些是我们要进行向量转换的原始文本
sentences = [   #传入的是列表
    'The weather is lovely today.',   # 天气真好
    'It is so sunny outside!',        # 外面阳光明媚
    'He drove to the stadium.',       # 他开车去了体育场
]

# 3.调用模型计算每个句子的嵌入向量，返回一个 numpy 数组，形状为 (句子数, 向量维度)
embeddings = model.encode(sentences)  
# 打印 embeddings 的类型
print(type(embeddings))  
# 打印嵌入向量的形状，应该是 (3, 384)
# 3 是你输入的句子数量（sentences 列表中有 3 个句子）。
# 384 是每个句子的 embedding 向量维度，也就是每个句子被表示成一个长度为 384 的向量。
print(embeddings.shape)

# 4. 打印每个句子的嵌入向量（向量内容），便于查看每个句子的表示
for i in embeddings:
    print(i)#向量内容
    print('--------------')

# 5. 编码一个新的单词 'sun'，也转换成嵌入向量，作为用户查询问题进行比对
new_embedding = model.encode(['sun'])  # 新的查询（可以理解为用户的问题）

# 相似度计算
# similarities = model.similarity(embeddings, embeddings)
similarities = model.similarity(new_embedding, embeddings)
# 输出一个 (1, 3) 的数组，每个值表示与一个句子的相似度
print(similarities)
```
